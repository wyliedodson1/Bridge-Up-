```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
csvFile = open("chlorophyll_mgm3_Dec_2002.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_list2002 = list(csvReader)
```

```python
#print(data_list2002[0])
```

```python
print(len(data_list2002[0]))
```

```python
print(len(data_list2002))
```

```python
width= len(data_list2002[0])
height= len(data_list2002)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8)
```

```python
all_value2002 = [] 

for r, row in enumerate(data_list2002):
    for c, value in enumerate(row):
        if float(value) < 99999:
            all_value2002.append(float(value)) 
        if float(value) <= 0.0111:
            grid[r, c] = [146, 179, 159]
        elif float(value) <= 0.05:
            grid[r, c] = [146, 179, 159]
        elif float(value) <= 0.1:
            grid[r, c] = [106, 156, 120]
        elif float(value) <= 0.15:
            grid[r, c] = [2, 227, 88]
        elif float(value) <= 0.25:
            grid[r, c] = [2, 199, 77]
        elif float(value) <= 0.5:
            grid[r, c] = [0, 166, 63]
        elif float(value) <= 10:
            grid[r, c] = [1, 112, 43]
        elif float(value) <= 53:
            grid[r, c]= [29, 105, 51]
        elif float(value) <= 70:
            grid[r, c] = [12, 82, 32]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(max(all_value2002))
print(min(all_value2002))
```

```python
plt.title("Chlorophyll in the Oceans; 2002 (mg/m^3)")
plt.imshow(grid)
plt.axis('off')
plt.scatter([], [], c= (146/255, 179/255, 159/255), label="Lowest Concentration")
plt.scatter([], [], c= (2/255, 227/255, 88/255), label="Medium Concentration")
plt.scatter([], [], c= (12/255, 82/255, 32/255), label="Highest Concentration")
plt.legend(bbox_to_anchor=(1.0, 0.03))
plt.savefig("Land_temperature_2002.png",dpi = 1000)
plt.show()
```
